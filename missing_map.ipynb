##Missing Map em Python
Uma maneira de analisar visualmente os dados missing do conjunto de dados é utilizar a biblioteca *missingno* para plotar um mapa dos dados faltantes. 
A seguir mostro o passo a passo para criar esse mapa dos missings data.

Para maiores informações da biblioteca *missingno* veja o github:https://github.com/ResidentMario/missingno

```
import pandas as pd
df=pd.read_csv('titanic_train.csv')
df
```

Importando a biblioteca *missigno* e gerando o gráfico missing map.

```
 import missingno as msno
%matplotlib inline
msno.matrix(df, color=(.07,.03,.5));
```

Observe que temos vários dados faltantes nas variáveis *age, cabin, boat, body* e *home*. A variável *fare* tem apenas observação faltante.
